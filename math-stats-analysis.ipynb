```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

#There are 3 buses on a certain route. The failures of the buses are independent and the probability of each bus failing in a day is 0.2

p = 0.2
n = 3
x = np.arange(0, n+1) # array of n+1 elements (+1 for 0 indexing meaning no failures)

# Probability of 0 failures in a day
probability = stats.binom.pmf(x[0],n,p)
print(round(probability, 2))

# Array with probabilities of failure occurence with indexes corresponding to the number of failures (in a day)
probabilities = stats.binom.pmf(x, n, p)
print(f"Possibility of breakdowns [0, 1, 2, 3] {probabilities}")

mean = n * p
print(f"Mean: {round(mean, 2)}")

variance =  mean * (1-p)
print(f"Variance: {round(variance, 2)}")

standard_deviation = np.sqrt(variance)
print(f"Standard deviation: {round(standard_deviation, 2)}")

#Expected value of days with no fails in 30 days
no_fails30 = 30 * probability
print(f"Probable days with no failures in a month: {round(no_fails30, 2)}")

# Probability function graph
plt.plot(x, probabilities, marker='o', color='blue', linestyle='-', alpha=0.7)
plt.xticks(x)
plt.xlabel('Number of failures')
plt.ylabel('Probability')
plt.title('Probability of number of the failures during one day')
plt.grid()
plt.show()

```
